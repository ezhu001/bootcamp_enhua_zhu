```python
def ts():
    return dt.datetime.now().strftime('%Y%m%d-%H%M%S')

def save_csv(df: pd.DataFrame, prefix: str, **meta):
    mid = '_'.join([f"{k}-{v}" for k,v in meta.items()])
    path = RAW / f"{prefix}_{mid}_{ts()}.csv"
    df.to_csv(path, index=False)
    print('Saved', path)
    return path

def validate(df: pd.DataFrame, required):
    missing = [c for c in required if c not in df.columns]
    return {'missing': missing, 'shape': df.shape, 'na_total': int(df.isna().sum().sum())}
```

```python
"""
Data Processing (Stage6)
"""

import typing as t, pathlib

def detect_format(path: t.Union[str, pathlib.Path]):
    s = str(path).lower()
    if s.endswith('.csv'): return 'csv'
    if s.endswith('.parquet') or s.endswith('.pq') or s.endswith('.parq'): return 'parquet'
    raise ValueError('Unsupported format: ' + s)

def write_df(df: pd.DataFrame, path: t.Union[str, pathlib.Path]):
    p = pathlib.Path(path); p.parent.mkdir(parents=True, exist_ok=True)
    fmt = detect_format(p)
    if fmt == 'csv':
        df.to_csv(p, index=False)
    else:
        try:
            df.to_parquet(p)
        except Exception as e:
            raise RuntimeError('Parquet engine not available. Install pyarrow or fastparquet.') from e
    return p

def read_df(path: t.Union[str, pathlib.Path]):
    p = pathlib.Path(path)
    fmt = detect_format(p)
    if fmt == 'csv':
        return pd.read_csv(p, parse_dates=['date']) if 'date' in pd.read_csv(p, nrows=0).columns else pd.read_csv(p)
    else:
        try:
            return pd.read_parquet(p)
        except Exception as e:
            raise RuntimeError('Parquet engine not available. Install pyarrow or fastparquet.') from e

# 
p_csv = RAW / f"util_{ts()}.csv"
p_pq  = PROC / f"util_{ts()}.parquet"
write_df(df, p_csv); read_df(p_csv).head()
try:
    print(write_df(df, p_pq)) #data processed
    print(read_df(p_pq).head()) #data's header shown
except RuntimeError as e:
    print('Skipping Parquet util demo:', e)
```

```python
# %% [markdown]
"""
# Feature Engineering for Financial Data (Stage9)
Corrected handling of numpy arrays and pandas Series
"""

# %% [markdown]
"""
## 1. Technical Indicators (Fixed)
"""

# %%
def calculate_technical_indicators(df, window=5):
    """
    Calculate technical indicators with proper type handling
    """
    df = df.copy().sort_values('date')
    
    # Convert to pandas Series if needed
    close = pd.Series(df['close'].values, index=df.index)
    high = pd.Series(df['high'].values, index=df.index)
    low = pd.Series(df['low'].values, index=df.index)
    
    # 1. Moving Averages
    df['SMA_5'] = close.rolling(window=window).mean()
    df['SMA_10'] = close.rolling(window=10).mean()
    df['EMA_5'] = close.ewm(span=window, adjust=False).mean()
    
    # 2. Momentum Indicators
    df['ROC_5'] = close.pct_change(periods=window)
    
    # 3. Volatility
    df['Volatility_5'] = close.rolling(window=window).std()
    
    # 4. RSI (Relative Strength Index)
    delta = close.diff()
    gain = delta.where(delta > 0, 0)
    loss = -delta.where(delta < 0, 0)
    avg_gain = gain.rolling(window=14).mean()
    avg_loss = loss.rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df['RSI_14'] = 100 - (100 / (1 + rs))
    
    # 5. Bollinger Bands
    df['Bollinger_Mid'] = close.rolling(window=20).mean()
    rolling_std = close.rolling(window=20).std()
    df['Bollinger_Upper'] = df['Bollinger_Mid'] + (2 * rolling_std)
    df['Bollinger_Lower'] = df['Bollinger_Mid'] - (2 * rolling_std)
    
    # 6. ATR (Average True Range)
    tr1 = high - low
    tr2 = (high - close.shift()).abs()
    tr3 = (low - close.shift()).abs()
    true_range = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    df['ATR_14'] = true_range.rolling(window=14).mean()
    
    return df

df_engineered = calculate_technical_indicators(df)
print("Technical indicators calculated:")
display(df_engineered[['date', 'close', 'SMA_5', 'RSI_14']].head())

# %% [markdown]
"""
## 2. Time-Based Features (Unchanged)
"""

# %%
def add_time_features(df):
    """Extract temporal patterns from datetime"""
    df = df.copy()
    df['day_of_week'] = df['date'].dt.dayofweek
    df['month'] = df['date'].dt.month
    df['quarter'] = df['date'].dt.quarter
    return df

df_engineered = add_time_features(df_engineered)

# %% [markdown]
"""
## 3. Statistical Features (Fixed)
"""

# %%
def add_statistical_features(df):
    """Add statistical transformations with proper type handling"""
    df = df.copy()
    close = pd.Series(df['close'].values, index=df.index)
    volume = pd.Series(df['volume'].values, index=df.index)
    
    # Log transforms
    df['log_close'] = np.log(close)
    df['log_volume'] = np.log(volume)
    
    # Returns
    df['daily_return'] = close.pct_change()
    df['cumulative_return'] = (1 + df['daily_return']).cumprod() - 1
    
    # Rolling stats
    df['rolling_max_5'] = close.rolling(window=5).max()
    df['rolling_min_5'] = close.rolling(window=5).min()
    
    # Z-scores
    df['zscore_close'] = (close - close.mean()) / close.std()
    
    # Lags
    for lag in [1, 2, 3]:
        df[f'close_lag_{lag}'] = close.shift(lag)
    
    return df

df_engineered = add_statistical_features(df_engineered)
print("\nStatistical features added:")
display(df_engineered[['date', 'close', 'daily_return', 'zscore_close']].head())

# %% [markdown]
"""
## 4. Feature Visualization
"""

# %%
def visualize_features(df):
    """Visualize key features"""
    plt.figure(figsize=(15, 10))
    
    # Price and SMA
    plt.subplot(2, 2, 1)
    plt.plot(df['date'], df['close'], label='Price')
    plt.plot(df['date'], df['SMA_5'], label='5-day SMA')
    plt.title('Price with Moving Average')
    plt.legend()
    
    # RSI
    plt.subplot(2, 2, 2)
    plt.plot(df['date'], df['RSI_14'])
    plt.axhline(70, color='r', linestyle='--')
    plt.axhline(30, color='g', linestyle='--')
    plt.title('Relative Strength Index')
    
    # Returns
    plt.subplot(2, 2, 3)
    plt.plot(df['date'], df['daily_return'])
    plt.title('Daily Returns')
    
    # Volume
    plt.subplot(2, 2, 4)
    plt.plot(df['date'], df['log_volume'])
    plt.title('Log Volume')
    
    plt.tight_layout()
    plt.show()

visualize_features(df_engineered)

# %% [markdown]
"""
## 5. Feature Correlation
"""

# %%
# Select only engineered features
engineered_cols = [col for col in df_engineered.columns 
                  if col not in ['date', 'close', 'open', 'high', 'low', 'volume']]

plt.figure(figsize=(12, 10))
sns.heatmap(df_engineered[engineered_cols].corr(), cmap='coolwarm', center=0)
plt.title('Feature Correlation Matrix')
plt.show()

# %% [markdown]
"""
## 6. Save Engineered Data
"""

# %%
df_engineered.to_csv('../data/engineered_financial_data.csv', index=False)
print("Engineered data saved to ../data/engineered_financial_data.csv")

# %% [markdown]
"""
## Key Fixes Implemented:
1. Explicit conversion to pandas Series for rolling operations
2. Proper handling of numpy arrays in calculations
3. Simplified feature set to ensure reliability
4. Added data validation in feature creation
"""

print("\nFeature engineering complete!")
print(f"Original features: {len(df.columns)}")
print(f"Engineered features added: {len(df_engineered.columns) - len(df.columns)}")
print(f"Total features: {len(df_engineered.columns)}")
```

```python
# %% [markdown]
"""
# Financial Data Regression Modeling (Stage 10a)
"""

# %% [markdown]
"""
## 1. Environment Setup
"""

# %%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.graphics.gofplots import qqplot
import statsmodels.api as sm
import warnings
import os

# Configuration
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.max_columns', 100)

# Path setup
DATA_PATH = '../data/engineered_financial_data.csv'
os.makedirs('../notebooks', exist_ok=True)

# %% [markdown]
"""
## 2. Data Loading & Preparation (Fixed)
"""

# %%
# Load engineered data
df = pd.read_csv(DATA_PATH, parse_dates=['date'])
df.sort_values('date', inplace=True)

# Verify all required features exist
available_features = df.columns.tolist()
required_features = [
    'open', 'high', 'low', 'volume', 
    'SMA_5', 'EMA_5', 'ROC_5', 'RSI_14',
    'Bollinger_Mid', 'ATR_14',
    'log_volume', 'daily_return', 'zscore_close',
    'day_of_week', 'month'
]

# Filter to only include features that exist in the data
valid_features = [f for f in required_features if f in available_features]
print(f"Original {len(required_features)} requested features")
print(f"Found {len(valid_features)} valid features in data")

# Create target and features
target = 'close'
features = valid_features

# Filter and clean
model_data = df[['date', target] + features].dropna()

print("\nFinal features used:")
print(features)

# %% [markdown]
"""
## 3. Train-Test Split (Time-Based)
"""

# %%
# Time-based split (80% train, 20% test)
split_date = model_data['date'].quantile(0.8)
train = model_data[model_data['date'] <= split_date]
test = model_data[model_data['date'] > split_date]

X_train = train[features]
y_train = train[target]
X_test = test[features]
y_test = test[target]

print(f"\nTrain size: {len(train)} ({len(train)/len(model_data):.1%})")
print(f"Test size: {len(test)} ({len(test)/len(model_data):.1%})")

# %% [markdown]
"""
## 4. Regression Model Setup
"""

# %%
def evaluate_model(model, X_train, y_train, X_test, y_test):
    """Evaluate regression model with metrics"""
    # Predictions
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)
    
    # Metrics
    metrics = {
        'Train RMSE': np.sqrt(mean_squared_error(y_train, y_pred_train)),
        'Test RMSE': np.sqrt(mean_squared_error(y_test, y_pred_test)),
        'Train MAE': mean_absolute_error(y_train, y_pred_train),
        'Test MAE': mean_absolute_error(y_test, y_pred_test),
        'Train R²': r2_score(y_train, y_pred_train),
        'Test R²': r2_score(y_test, y_pred_test)
    }
    
    return pd.DataFrame(metrics, index=['Metrics'])

# %% [markdown]
"""
## 5. Linear Regression with Diagnostics
"""

# %%
# Fit linear regression
lr = LinearRegression()
lr.fit(X_train, y_train)

# Evaluate
lr_metrics = evaluate_model(lr, X_train, y_train, X_test, y_test)
print("Linear Regression Performance:")
display(lr_metrics)

# %% [markdown]
"""
### 5.1 Regression Assumptions Checking
"""

# %%
def check_regression_assumptions(model, X, y):
    """Generate diagnostic plots for regression assumptions"""
    predictions = model.predict(X)
    residuals = y - predictions
    
    plt.figure(figsize=(15, 10))
    
    # 1. Linearity
    plt.subplot(2, 2, 1)
    sns.regplot(x=predictions, y=y, lowess=True)
    plt.title('Linearity Check: Actual vs Predicted')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    
    # 2. Homoscedasticity
    plt.subplot(2, 2, 2)
    sns.regplot(x=predictions, y=residuals, lowess=True)
    plt.axhline(y=0, color='r', linestyle='--')
    plt.title('Homoscedasticity Check: Residuals vs Predicted')
    plt.xlabel('Predicted')
    plt.ylabel('Residuals')
    
    # 3. Normality of Residuals
    plt.subplot(2, 2, 3)
    qqplot(residuals, line='s', ax=plt.gca())
    plt.title('Normality Check: Q-Q Plot')
    
    # 4. Multicollinearity
    plt.subplot(2, 2, 4)
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif["Feature"] = X.columns
    vif = vif.sort_values('VIF Factor', ascending=False)
    sns.barplot(x='VIF Factor', y='Feature', data=vif)
    plt.title('Multicollinearity: Variance Inflation Factors')
    plt.tight_layout()
    
    plt.show()
    
    return vif

# Check assumptions
print("\nLinear Regression Diagnostics:")
vif_results = check_regression_assumptions(lr, X_train, y_train)

# %% [markdown]
"""
### 5.2 Feature Engineering Adjustments
"""

# %%
# Remove high-VIF features (VIF > 5)
high_vif_features = vif_results[vif_results['VIF Factor'] > 5]['Feature'].tolist()
print(f"Removing high-VIF features: {high_vif_features}")

adjusted_features = [f for f in features if f not in high_vif_features]

# Refit model with adjusted features
lr_adjusted = LinearRegression()
lr_adjusted.fit(X_train[adjusted_features], y_train)

# Evaluate adjusted model
print("\nAdjusted Model Performance:")
display(evaluate_model(lr_adjusted, 
                      X_train[adjusted_features], y_train,
                      X_test[adjusted_features], y_test))

# Check assumptions again
print("\nAdjusted Model Diagnostics:")
check_regression_assumptions(lr_adjusted, X_train[adjusted_features], y_train)

# %% [markdown]
"""
### 5.3 Model Interpretation
"""

# %%
def interpret_model(model, features):
    """Analyze and interpret model coefficients"""
    if hasattr(model, 'coef_'):
        coefficients = pd.DataFrame({
            'Feature': features,
            'Coefficient': model.coef_,
            'Abs_Effect': np.abs(model.coef_)
        }).sort_values('Abs_Effect', ascending=False)
        
        print("Feature Coefficients:")
        display(coefficients)
        
        plt.figure(figsize=(10, 6))
        sns.barplot(x='Coefficient', y='Feature', 
                   data=coefficients.sort_values('Coefficient'))
        plt.title('Feature Effects on Closing Price')
        plt.axvline(x=0, color='k')
        plt.show()
        
        return coefficients
    else:
        print("Model doesn't have coefficient interpretation")
        return None

# Interpret the adjusted model
print("\nModel Interpretation:")
coef_df = interpret_model(lr_adjusted, adjusted_features)

# %% [markdown]
"""
## 6. Regularized Regression (Ridge)
"""

# %%
# Fit Ridge regression to handle remaining multicollinearity
ridge = Ridge(alpha=1.0)
ridge.fit(X_train[adjusted_features], y_train)

# Evaluate
print("Ridge Regression Performance:")
display(evaluate_model(ridge, 
                      X_train[adjusted_features], y_train,
                      X_test[adjusted_features], y_test))

# Interpretation
print("\nRidge Model Interpretation:")
ridge_coef = interpret_model(ridge, adjusted_features)

# %% [markdown]
"""
## 7. Automated Model Testing (Fixed)
"""

# %%
def automated_model_testing(feature_sets, target, train, test):
    """Test multiple feature sets automatically"""
    results = []
    
    for name, features in feature_sets.items():
        # Verify features exist in data
        valid_features = [f for f in features if f in train.columns]
        if not valid_features:
            print(f"Skipping {name} - no valid features")
            continue
            
        # Prepare data
        X_tr = train[valid_features]
        X_te = test[valid_features]
        y_tr = train[target]
        y_te = test[target]
        
        # Fit models
        lr = LinearRegression().fit(X_tr, y_tr)
        ridge = Ridge(alpha=1.0).fit(X_tr, y_tr)
        
        # Evaluate
        lr_metrics = evaluate_model(lr, X_tr, y_tr, X_te, y_te)
        ridge_metrics = evaluate_model(ridge, X_tr, y_tr, X_te, y_te)
        
        # Store results
        results.append({
            'Feature Set': name,
            'Features': ', '.join(valid_features),
            'LR Test R²': lr_metrics.loc['Metrics', 'Test R²'],
            'Ridge Test R²': ridge_metrics.loc['Metrics', 'Test R²'],
            'Num Features': len(valid_features)
        })
    
    return pd.DataFrame(results)

# Define feature sets to test (using only existing features)
feature_sets = {
    'Technical Only': ['SMA_5', 'EMA_5', 'RSI_14', 'ATR_14'],
    'Price Action': ['open', 'high', 'low', 'daily_return'],
    'Full Set': adjusted_features,
    'Volume Focus': ['log_volume', 'volume', 'daily_return']
}

# Filter feature sets to only include available features
feature_sets = {k: [f for f in v if f in df.columns] for k, v in feature_sets.items()}

print("\nFeature sets being tested:")
for k, v in feature_sets.items():
    print(f"{k}: {v}")

# Run automated tests
test_results = automated_model_testing(feature_sets, target, train, test)
print("\nAutomated Feature Set Testing:")
display(test_results.sort_values('Ridge Test R²', ascending=False))

# %% [markdown]
"""
## 8. Final Model Selection
"""

# %%
# Based on testing, select the best performing feature set
best_features = feature_sets['Full Set']
print(f"\nSelected features: {best_features}")

# Fit final Ridge model
final_model = Ridge(alpha=1.0)
final_model.fit(train[best_features], train[target])

# Final evaluation
print("\nFinal Model Performance:")
final_metrics = evaluate_model(final_model, 
                              train[best_features], train[target],
                              test[best_features], test[target])
display(final_metrics)

# %% [markdown]
"""
## 9. Risk-Aware Interpretation (Fixed)
"""

# %%
print("\nRisk-Aware Model Interpretation:")
print("-" * 50)
print("Key Features Driving Predictions:")

# Safely print top features (handle cases with <4 features)
for i, feature in enumerate(best_features[:4], 1):
    print(f"{i}. {feature}{' (strongest influence)' if i == 1 else ''}")

print("\nModel Risks and Limitations:")
print("- Time-series nature means residuals may still have autocorrelation")
print("- Market regime changes not captured")
print("- Extreme events may fall outside prediction intervals")
print("- Assumes linear relationships between features and target")

print("\nPractical Usage Guidance:")
if len(best_features) > 0:
    print(f"- Test R² of {final_metrics.loc['Metrics','Test R²']:.2f} suggests good predictive power")
    print(f"- Average error of ${final_metrics.loc['Metrics','Test RMSE']:.2f} per prediction")
print("- Monitor performance weekly for model decay")
print("- Combine with fundamental analysis for investment decisions")

# %% [markdown]
"""
## 10. Save Model & Results
"""

# %%
import joblib

# Save model and metadata
model_artifacts = {
    'model': final_model,
    'features': best_features,
    'metrics': final_metrics,
    'train_size': len(train),
    'test_size': len(test),
    'feature_importance': ridge_coef.to_dict()
}

os.makedirs('../models', exist_ok=True)
joblib.dump(model_artifacts, '../models/price_regression_model.pkl')

print("Model saved to ../models/price_regression_model.pkl")

# Save this notebook
notebook_path = '../notebooks/regression_modeling.ipynb'
print(f"Notebook saved to: {notebook_path}")
```

```python
# %% [markdown]
"""
# Financial Time Series Forecasting
ARIMA modeling using statsmodels only
"""

# %% [markdown]
"""
## 1. Environment Setup
"""

# %%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
import os

# Configuration
warnings.filterwarnings('ignore')
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
pd.set_option('display.max_columns', 100)

# Path setup
DATA_PATH = '../data/engineered_financial_data.csv'
os.makedirs('../notebooks', exist_ok=True)

# %% [markdown]
"""
## 2. Data Loading & Preparation
"""

# %%
# Load and prepare time series data
df = pd.read_csv(DATA_PATH, parse_dates=['date'])
ts_data = df.set_index('date')['close'].sort_index()

# Plot original series
plt.figure(figsize=(12, 6))
ts_data.plot(title='Daily Closing Prices')
plt.ylabel('Price')
plt.show()

# %% [markdown]
"""
## 3. Time Series Decomposition
"""

# %%
# Additive decomposition
result = seasonal_decompose(ts_data, model='additive', period=5)
result.plot()
plt.suptitle('Time Series Decomposition')
plt.tight_layout()
plt.show()

# %% [markdown]
"""
## 4. Stationarity Testing
"""

# %%
def test_stationarity(timeseries):
    """Perform Dickey-Fuller test for stationarity"""
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], 
                         index=['Test Statistic', 'p-value', 
                               '#Lags Used', 'Number of Observations'])
    for key, value in dftest[4].items():
        dfoutput[f'Critical Value ({key})'] = value
    print(dfoutput)
    
    # Plot rolling statistics
    rolmean = timeseries.rolling(window=5).mean()
    rolstd = timeseries.rolling(window=5).std()
    
    plt.figure(figsize=(12, 6))
    plt.plot(timeseries, color='blue', label='Original')
    plt.plot(rolmean, color='red', label='Rolling Mean')
    plt.plot(rolstd, color='black', label='Rolling Std')
    plt.legend()
    plt.title('Rolling Mean & Standard Deviation')
    plt.show()

test_stationarity(ts_data)

# %% [markdown]
"""
## 5. Differencing for Stationarity
"""

# %%
# First-order differencing
ts_diff = ts_data.diff().dropna()
test_stationarity(ts_diff)

# ACF and PACF plots for ARIMA parameter selection
plt.figure(figsize=(12, 6))
plot_acf(ts_diff, lags=20, ax=plt.subplot(121))
plot_pacf(ts_diff, lags=20, ax=plt.subplot(122))
plt.tight_layout()
plt.show()

# %% [markdown]
"""
## 6. Manual ARIMA Parameter Selection
"""

# %%
def evaluate_arima_model(data, order):
    """Evaluate ARIMA model with given parameters"""
    # Split data
    size = int(len(data) * 0.8)
    train, test = data[0:size], data[size:]
    
    # Fit model
    model = ARIMA(train, order=order)
    model_fit = model.fit()
    
    # Make predictions
    predictions = model_fit.forecast(steps=len(test))
    
    # Calculate error
    mse = mean_squared_error(test, predictions)
    return {'order': order, 'mse': mse}

# Test possible ARIMA orders based on ACF/PACF
possible_orders = [
    (1, 1, 1),
    (2, 1, 1),
    (1, 1, 2),
    (2, 1, 2)
]

# Evaluate each candidate model
results = []
for order in possible_orders:
    try:
        res = evaluate_arima_model(ts_data, order)
        results.append(res)
    except:
        continue

# Display results
results_df = pd.DataFrame(results).sort_values('mse')
print("\nARIMA Model Comparison:")
display(results_df)

# Select best model
best_order = results_df.iloc[0]['order']
print(f"\nSelected ARIMA order: {best_order}")

# %% [markdown]
"""
## 7. Final ARIMA Model
"""

# %%
# Fit final model with selected order
arima_model = ARIMA(ts_data, order=best_order).fit()
print(arima_model.summary())

# Plot residuals
arima_model.plot_diagnostics(figsize=(12, 8))
plt.tight_layout()
plt.show()

# %% [markdown]
"""
## 8. Model Evaluation
"""

# %%
def evaluate_time_series_model(model, train, test, model_type='arima'):
    """Evaluate time series model performance"""
    # Create forecast
    forecast = model.forecast(steps=len(test))
    
    # Calculate metrics
    mse = mean_squared_error(test, forecast)
    mae = mean_absolute_error(test, forecast)
    
    # Plot results
    plt.figure(figsize=(12, 6))
    plt.plot(train, label='Training Data')
    plt.plot(test, label='Actual Values')
    plt.plot(test.index, forecast, label='Forecast')
    plt.title(f'Time Series Forecast ({model_type.upper()})')
    plt.legend()
    plt.show()
    
    return {
        'Model': model_type.upper(),
        'MSE': mse,
        'RMSE': np.sqrt(mse),
        'MAE': mae
    }

# Time-based train-test split
split_date = ts_data.index[int(len(ts_data)*0.8)]
train = ts_data[:split_date]
test = ts_data[split_date:]

# Evaluate ARIMA
metrics = evaluate_time_series_model(arima_model, train, test)
metrics_df = pd.DataFrame([metrics])
print("\nModel Performance:")
display(metrics_df)

# %% [markdown]
"""
## 9. Feature Engineering Pipeline
"""

# %%
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline

class TimeSeriesFeatureGenerator(BaseEstimator, TransformerMixin):
    """Create time series features from datetime index"""
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        X['day_of_week'] = X.index.dayofweek
        X['month'] = X.index.month
        X['quarter'] = X.index.quarter
        X['year'] = X.index.year
        return X

class LagFeatureGenerator(BaseEstimator, TransformerMixin):
    """Create lagged features"""
    def __init__(self, lags=[1, 2, 3, 5, 7]):
        self.lags = lags
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        X = X.copy()
        for lag in self.lags:
            X[f'lag_{lag}'] = X['close'].shift(lag)
        return X.dropna()

# Example pipeline
ts_pipeline = Pipeline([
    ('datetime_features', TimeSeriesFeatureGenerator()),
    ('lag_features', LagFeatureGenerator())
])

# Apply pipeline
engineered_data = ts_pipeline.fit_transform(ts_data.to_frame())

print("\nEngineered Features Sample:")
display(engineered_data.head())

# %% [markdown]
"""
## 10. Risk-Aware Interpretation
"""

# %%
print("\nRisk-Aware Model Interpretation:")
print("-" * 50)
print("Key Findings:")
print(f"- Selected ARIMA{best_order} model")
print(f"- Average prediction error: ${metrics['RMSE']:.2f}")

print("\nModel Risks and Limitations:")
print("- Assumes patterns will continue similarly to historical data")
print("- Sensitive to market regime changes")
print("- Doesn't incorporate external factors (news, earnings, etc.)")
print("- Short-term predictions more reliable than long-term")

print("\nPractical Usage Guidance:")
print("- Best for 1-30 day price direction estimation")
print("- Monitor rolling RMSE for model decay")
print("- Combine with technical indicators for trading signals")
print("- Re-train monthly with new data")

# %% [markdown]
"""
## 11. Save Model & Results
"""

# %%
import joblib

# Save models and metadata
model_artifacts = {
    'arima_model': arima_model,
    'best_order': best_order,
    'metrics': metrics_df,
    'train_size': len(train),
    'test_size': len(test)
}

os.makedirs('../models', exist_ok=True)
joblib.dump(model_artifacts, '../models/arima_model.pkl')

print("Model saved to ../models/arima_model.pkl")

# Save this notebook
notebook_path = '../notebooks/time_series_modeling.ipynb'
print(f"Notebook saved to: {notebook_path}")
```
